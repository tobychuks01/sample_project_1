```python
#Import Libraries (In Jupyter)
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Create Columns
columns = [
    "Date_posted",
    "Video_Title/hook",
    "Video_Type",
    "Platform",
    "Views",
    "Likes",
    "Shares",
    "Saves",
    "Comments",
    "Completion_rate(%)",
    "CTR",
    "Engagement",
    "Notes"
]

df = pd.DataFrame(columns=columns)

#Fill columns with data(Always add data to calculate)
data = [
    ["2026-02-01", "Nobody tells you this about trading", "Curiosity", "TikTok", 12000, 1400, 300,300, 250, 180, 72, 3.2, ""],
    ["2026-02-02", "POV: You finally understand liquidity", "POV", "Instagram", 8000, 600, 120, 120, 90, 70, 58, 2.1, ""],
    ["2026-02-03", "Why 90% of traders lose money", "Authority", "TikTok", 20000, 2500, 800, 800, 700, 600, 75, 5.4, ""],
    ["2026-02-18", "Why Toby chuks is gonaa become a Billionaire", "Authority", "TikTok", 100000, 400000, 3003939,3003939,9398383,8383838,1000,10.0,""]
]

df = pd.DataFrame(data, columns=columns)

#Convert Date Column
df["Date_posted"] = pd.to_datetime(df["Date_posted"])

#Make Sure Numbers Are Numbers
numeric_cols = [
    "Views","Likes","Shares","Saves","Comments",
    "Completion_rate(%)","CTR"
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)

#Calculate Engagement 
df["Engagement"] = (
    df["Likes"]
    + df["Shares"] * 2
    + df["Saves"] * 3
    + df["Comments"] * 2
)

#Calculate Engagement Score 
#(Formula(Engagement = (Likes + Shares*2 + Saves*3 + Comments*2) / Views * 100))
df["engagement_score"] = (
    (
        df["Likes"] +
        df["Shares"]*2 +
        df["Saves"]*3 +
        df["Comments"]*2
    ) / df["Views"]
) * 100



#Round Values
df["engagement_score"] = df["engagement_score"].round(2)

#Auto Assign Status (KILL / SCALE / ITERATE)
#Using NumPy Conditions
conditions = [
    (df["engagement_score"] > 8) & (df["Completion_rate(%)"] > 60),
    (df["engagement_score"] < 3) & (df["Completion_rate(%)"] < 40)
]

choices = ["SCALE", "KILL"]

df["status"] = np.select(conditions, choices, default="ITERATE")

#Weekly Insights Analysis
#Top Performing Videos
top_videos = df.sort_values("engagement_score", ascending=False)
top_videos[["Video_Title/hook","engagement_score"]]

#Best Video Type
Best_video_type = df.groupby("Video_Title/hook")["engagement_score"].mean().sort_values(ascending=False)

#Best Platform
Best_platform = df.groupby("Platform")["engagement_score"].mean()


#Visualization (Matplotlib)
#Engagement Score Per Video plt.figure()
#plt.bar(df["Video_Title/hook"], df["engagement_score"])
#plt.xticks(rotation=10)
#plt.title("Engagement Score Per Video")
#plt.ylabel("Engagement %")
#plt.show();

#Views vs Engagement Relationship
#plt.figure()
#plt.scatter(df["Views"], df["engagement_score"])
#plt.title("Views vs Engagement")
#plt.xlabel("Views")
#plt.ylabel("Engagement Score")
#plt.show();

#Monthly Filters (Decision Layer)
#Scale Videos
#scale_video = df[df["engagement_score"] > 10]
#scale_video

#Bar chart for metrics
#metrics = ["Views", "Likes", "Shares", "Saves", "Comments"]
# Sum totals across all videos
#metric_totals = df[metrics].sum()
# Convert to DataFrame for plotting (optional but clean)
#metrics_df = metric_totals.reset_index()
#metrics_df.columns = ["Metric", "Total"]
# Plot bar chart
#plt.figure()
#plt.bar(metrics_df["Metric"], metrics_df["Total"])
#plt.title("Total Performance Metrics Comparison")
#plt.xlabel("Metrics")
#plt.ylabel("Total Count")
#plt.show()

#Per video
#metrics = ["Views", "Likes", "Shares", "Saves", "Comments"]

#df.set_index("Video_Title/hook")[metrics].plot(kind="bar")

#plt.title("Metrics Per Video")
#plt.xlabel("Video")
#plt.ylabel("Count")
#plt.xticks(rotation=45)
#plt.show()

#Videos to kill
kill_df = df[df["engagement_score"] < 3]
kill_df

df.to_csv("Cipher.csv", index=False)

```
