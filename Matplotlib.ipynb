```python
butter_sales = {"ALMOND BUTTER": 10,
                "PEANUT BUTTER": 8,
                "CASHEW BUTTER": 12}
fig, ax = plt.subplots()
ax.bar( butter_sales.keys(), butter_sales.values())
ax.set(title = "Toby's Butter store",
       ylabel = " Price ($)");
fig.savefig("imagesplt/Toby's-butter-sales.png")
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2,
                                             ncols= 2,
                                             figsize= (15,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.rand(10), np.random.rand(10));
ax3.bar(butter_sales.keys(), butter_sales.values());
ax4.hist(np.random.randint(1000));
ax.set(title = "Random Plots");
fig.savefig("imagesplt/Random-plots.png")
       
```
